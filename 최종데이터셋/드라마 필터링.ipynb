```python
df_cast_crew_filtered = df_cast_crew[df_cast_crew['imdb_id'].isin(df_dramas['imdb_id'])]
df_dramas_filtered = df_dramas[df_dramas['imdb_id'].isin(df_cast_crew['imdb_id'])] #(3890, 23) / (3891, 32)
print(df_cast_crew_filtered.shape)
df_dramas_filtered.shape
```

```python
df_cast_crew_filtered = df_cast_crew_filtered.drop(
    ['series_id', 'title', 'original_name', 'first_air_date',
     'vote_average', 'vote_count', 'popularity'],
    axis=1
)
```

```python
df_merged = df_dramas_filtered.merge(
    df_cast_crew_filtered,
    on="imdb_id",
    how="left"
)
df_merged.shape
```

```python
df_merged.to_parquet("tv_dramas_full.parquet", engine="fastparquet", index=False)
```
